# 图片比较

## 1. 灰度比较

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

# load images
image1 = cv2.imread("images/image1.jpg")
image2 = cv2.imread("images/image1_diff.jpg")

# compute difference
diff = cv2.subtract(image1, image2)

plt.imshow(diff)
plt.title('input')
plt.show()
```
